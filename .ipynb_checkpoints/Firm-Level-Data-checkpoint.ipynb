## Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

## Loading the Dataset

```python
data = pd.read_csv("Firm_level_data.csv")
```

```python
data.head()
```

```python
# Checking null values
print(data.isnull().sum())
```

```python
print(data.info())
print(data.shape)
```

```python
print(data.describe())
```

```python
# Univariate Analysis
sns.histplot(data['sales'], kde=True)
plt.title('Distribution of Sales')
plt.show()
```

```python
# Bivariate Analysis
sns.pairplot(data)
plt.show()
```

## Handling Null Values

```python
non_numeric_columns = []
for column in data.columns:
    if data[column].dtype == 'object':
        non_numeric_columns.append(column)
```

```python
print("Columns with non-numeric values:", non_numeric_columns)
```

```python
for column in non_numeric_columns:
    unique_values = data[column].unique()
    print(f"Unique values in {column}:", unique_values)
```

```python
data.replace({'yes': 1, 'no': 0}, inplace=True)
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
X = data.drop(columns=['No.', 'sales'])
```

```python
y = data['sales']
```

## Splitting the Data into Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Applying Linear Regression

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

## Predictions

```python
train_preds = model.predict(X_train)
```

```python
test_preds = model.predict(X_test)
```

## Performance Metrics

```python
train_r2 = r2_score(y_train, train_preds)
```

```python
test_r2 = r2_score(y_test, test_preds)
```

```python
train_rmse = np.sqrt(mean_squared_error(y_train, train_preds))
```

```python
test_rmse = np.sqrt(mean_squared_error(y_test, test_preds))
```

```python
# Interpretation of Model Performance
print("Linear Regression Model Performance:")
print("Train R-squared:", train_r2*100)
print("Test R-squared:", test_r2*100)
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
```

```python
coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.coef_})
```

```python
coefficients = coefficients.sort_values(by='Coefficient', ascending=False)
```

```python
top_5_attributes = coefficients.head(5)
```

```python
print("Top 5 Important Attributes:")
print(top_5_attributes)
```
